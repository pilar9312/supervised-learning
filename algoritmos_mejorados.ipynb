Al pasar los datos a python tenemos la siguiente implementacion

```python
dataset = [
    dict(
        dolor_garganta = True,
        congestion_nasal = False,
        fiebre = True,
        dificultad_respirar = True, 
        covid = True        
    ),
    dict(
        dolor_garganta = True,
        congestion_nasal = True,
        fiebre = False,
        dificultad_respirar = False, 
        covid = False        
    ),
    dict(
        dolor_garganta = True,
        congestion_nasal = False,
        fiebre = True,
        dificultad_respirar = True, 
        covid = True        
    ),
    dict(
        dolor_garganta = False,
        congestion_nasal = True,
        fiebre = True,
        dificultad_respirar = True, 
        covid = True        
    ),
    dict(
        dolor_garganta = False,
        congestion_nasal = False,
        fiebre = False,
        dificultad_respirar = False, 
        covid = False        
    ),
    dict(
        dolor_garganta = False,
        congestion_nasal = True,
        fiebre = True,
        dificultad_respirar = True, 
        covid = True        
    ),
    dict(
        dolor_garganta = False,
        congestion_nasal = True,
        fiebre = False,
        dificultad_respirar = False, 
        covid = False        
    ),
    dict(
        dolor_garganta = True,
        congestion_nasal = True,
        fiebre = True,
        dificultad_respirar = True, 
        covid = True        
    )
]
```

- Se calculan probabilidades para cada sintoma

```python
p_covid = len([item for item in dataset if item["covid"]])/len(dataset)
print(f"p_covid = {p_covid}")
p_no_covid = len([item for item in dataset if not item["covid"]])/len(dataset)
print(f"p_no_covid = {p_no_covid}")

# Calculo probabilidades covid = True
p_dolor_garganta_given_covid = len([item for item in dataset if item["dolor_garganta"] and item["covid"]])/(len(dataset)*p_covid)
print(f"p_dolor_garganta_given_covid = {p_dolor_garganta_given_covid}")
p_no_dolor_garganta_given_covid = len([item for item in dataset if not item["dolor_garganta"] and item["covid"]])/(len(dataset)*p_covid)
print(f"p_no_dolor_garganta_given_covid = {p_no_dolor_garganta_given_covid}")
p_congestion_nasal_given_covid = len([item for item in dataset if item["congestion_nasal"] and item["covid"]])/(len(dataset)*p_covid)
print(f"p_congestion_nasal_given_covid = {p_congestion_nasal_given_covid}")
p_no_congestion_nasal_given_covid = len([item for item in dataset if not item["congestion_nasal"] and item["covid"]])/(len(dataset)*p_covid)
print(f"p_no_congestion_nasal_given_covid = {p_no_congestion_nasal_given_covid}")
p_fiebre_given_covid = len([item for item in dataset if item["fiebre"] and item["covid"]])/(len(dataset)*p_covid)
print(f"p_fiebre_given_covid = {p_fiebre_given_covid}")
p_no_fiebre_given_covid = len([item for item in dataset if not item["fiebre"] and item["covid"]])/(len(dataset)*p_covid)
print(f"p_no_fiebre_given_covid = {p_no_fiebre_given_covid}")
p_dificultad_respirar_given_covid = len([item for item in dataset if item["dificultad_respirar"] and item["covid"]])/(len(dataset)*p_covid)
print(f"p_dificultad_respirar_given_covid = {p_dificultad_respirar_given_covid}")
p_no_dificultad_respirar_given_covid = len([item for item in dataset if not item["dificultad_respirar"] and item["covid"]])/(len(dataset)*p_covid)
print(f"p_no_dificultad_respirar_given_covid = {p_no_dificultad_respirar_given_covid}")

# calculo probabilidades covid = False
p_dolor_garganta_given_no_covid = len([item for item in dataset if item["dolor_garganta"] and not item["covid"]])/(len(dataset)*p_no_covid)
print(f"p_dolor_garganta_given_no_covid = {p_dolor_garganta_given_no_covid}")
p_no_dolor_garganta_given_no_covid = len([item for item in dataset if not item["dolor_garganta"] and not item["covid"]])/(len(dataset)*p_no_covid)
print(f"p_no_dolor_garganta_given_no_covid = {p_no_dolor_garganta_given_no_covid}")
p_congestion_nasal_given_no_covid = len([item for item in dataset if item["congestion_nasal"] and not item["covid"]])/(len(dataset)*p_no_covid)
print(f"p_congestion_nasal_given_no_covid = {p_congestion_nasal_given_no_covid}")
p_no_congestion_nasal_given_no_covid = len([item for item in dataset if not item["congestion_nasal"] and not item["covid"]])/(len(dataset)*p_no_covid)
print(f"p_no_congestion_nasal_given_no_covid = {p_no_congestion_nasal_given_no_covid}")
p_fiebre_given_no_covid = len([item for item in dataset if item["fiebre"] and not item["covid"]])/(len(dataset)*p_no_covid)
print(f"p_fiebre_given_no_covid = {p_fiebre_given_no_covid}")
p_no_fiebre_given_no_covid = len([item for item in dataset if not item["fiebre"] and not item["covid"]])/(len(dataset)*p_no_covid)
print(f"p_no_fiebre_given_no_covid = {p_no_fiebre_given_no_covid}")
p_dificultad_respirar_given_no_covid = len([item for item in dataset if item["dificultad_respirar"] and not item["covid"]])/(len(dataset)*p_no_covid)
print(f"p_dificultad_respirar_given_no_covid = {p_dificultad_respirar_given_no_covid}")
p_no_dificultad_respirar_given_no_covid = len([item for item in dataset if not item["dificultad_respirar"] and not item["covid"]])/(len(dataset)*p_no_covid)
print(f"p_no_dificultad_respirar_given_no_covid = {p_no_dificultad_respirar_given_no_covid}")
```

- Se calculan las probabilidades de todos los sintomas

```python
p_sintomas_given_covid = p_dolor_garganta_given_covid * p_congestion_nasal_given_covid * p_fiebre_given_covid * p_dificultad_respirar_given_covid
print(f"p_sintomas_given_covid = {p_sintomas_given_covid}")
p_no_sintomas_given_covid = p_no_dolor_garganta_given_covid * p_no_congestion_nasal_given_covid * p_no_fiebre_given_covid * p_no_dificultad_respirar_given_covid
print(f"p_no_sintomas_given_covid = {p_no_sintomas_given_covid}")
p_sintomas_given_no_covid = p_dolor_garganta_given_no_covid * p_congestion_nasal_given_no_covid * p_fiebre_given_no_covid * p_dificultad_respirar_given_no_covid
print(f"p_sintomas_given_no_covid = {p_sintomas_given_no_covid}")
p_no_sintomas_given_no_covid = p_no_dolor_garganta_given_no_covid * p_no_congestion_nasal_given_no_covid * p_no_fiebre_given_no_covid * p_no_dificultad_respirar_given_no_covid
print(f"p_no_sintomas_given_no_covid = {p_no_sintomas_given_no_covid}")
p_sintomas = p_sintomas_given_covid * p_covid + p_sintomas_given_no_covid * p_no_covid
print(f"p_sintomas = {p_sintomas}")
p_no_sintomas = p_no_sintomas_given_covid * p_covid + p_no_sintomas_given_no_covid * p_no_covid
print(f"p_no_sintomas = {p_no_sintomas}")
p_covid_given_sintomas = p_sintomas_given_covid * p_covid / p_sintomas
print(f"La probabilidad de que el paciente tenga COVID-19 dados los sintomas es {p_covid_given_sintomas}")
p_no_covid_given_sintomas = p_sintomas_given_no_covid * p_no_covid / p_sintomas
print(f"La probabilidad de que el paciente no tenga COVID-19 dados los sintomas es {p_no_covid_given_sintomas}")
```

- Calculo de probabilidades solo si se considera el sintoma dolor garganta

```python
p_dolor_garganta = len([item for item in dataset if item["dolor_garganta"]])/len(dataset)
print(f"p_dolor_garganta = {p_dolor_garganta}")
p_no_dolor_garganta = len([item for item in dataset if not item["dolor_garganta"]])/len(dataset)
print(f"p_no_dolor_garganta = {p_no_dolor_garganta}")
p_covid_given_dolor_garganta = p_dolor_garganta_given_covid * p_covid / p_dolor_garganta
print(f"La probabilidad de que el paciente tenga COVID-19 dado dolor de garganta es {p_covid_given_dolor_garganta}")
p_no_covid_given_sintomas = p_dolor_garganta_given_no_covid * p_no_covid / p_dolor_garganta
print(f"La probabilidad de que el paciente no tenga COVID-19 dados dolor de garganta es {p_no_covid_given_sintomas}")
```

- Utilizacion del dataset para el arbol de decisiones

```python
import pandas as pd
df = pd.DataFrame(dataset) 
print(df)
df.dolor_garganta = df.dolor_garganta.replace({True: 1, False: 0})
df.congestion_nasal = df.congestion_nasal.replace({True: 1, False: 0})
df.fiebre = df.fiebre.replace({True: 1, False: 0})
df.dificultad_respirar = df.dificultad_respirar.replace({True: 1, False: 0})
df.covid = df.covid.replace({True: 1, False: 0})
df
```

```python
from matplotlib import pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split

# Se seleccionan las columnas para X y Y
X=df.iloc[:,:-1]
y=df.iloc[:,[4]]
print(df.columns)

# Convertir variables categricas en variables numricas
X = pd.get_dummies(X)

# Se escogio un 20% como datos de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Se entrena un modelo de arbol de decision
profundidad_arbol = 3
clf=DecisionTreeClassifier(max_depth=profundidad_arbol)
clf.fit(X_train, y_train)

# Evaluar el modelo con el conjunto de datos de prueba
score = clf.score(X_test, y_test)
print("Accuracy: %.2f%%" % (score*100.0))

# Verifiacion del modelo entrenado
y_pred = clf.predict(X_test)

# Se grafica el arbol y se guarda archivo de la grafica
clf.fit(X,y)
plt.figure(figsize=(80,20))
grafico = tree.plot_tree(decision_tree=clf, feature_names=X.columns, filled=True) 
plt.savefig("output.png")
```
