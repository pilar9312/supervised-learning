```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

#se cargan los datos del archivo
df = pd.read_csv("Cleaned-Data.csv")

# Eliminar filas con valores faltante
df.dropna(inplace=True)

# Se descartan para X las columnas que no seran usadas y se suman las severidades como y
X=df.iloc[:,[0,1,2,3,4,6,7,8,9]]
df['Severity'] = df['Severity_Mild'] + df['Severity_Severe'] + df['Severity_Moderate']
y=df.iloc[:,[27]]
print(f"Columnas seleccionadas para X:\n{X.columns}")

# Convertir variables categóricas en variables numéricas
X = pd.get_dummies(X)

# Dividir el conjunto de datos en conjunto de entrenamiento y pruebas
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values.ravel(), test_size=0.1, random_state=42)

# Crear el clasificador knn con k=4
knn = KNeighborsClassifier(n_neighbors=4)

# Entrenar el clasificador
knn.fit(X_train, y_train)

# Verifiacion del modelo entrenado
y_pred = knn.predict(X_test)

# Calcular precison del clasificador
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy: %.2f%%" % (accuracy*100.0))
```
